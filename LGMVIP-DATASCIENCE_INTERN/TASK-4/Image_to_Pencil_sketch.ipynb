Importing Libraries

```python
import matplotlib.pyplot as plt
%matplotlib inline
import cv2
```

Reading the original image

```python

image=cv2.imread("tiger.jpg")
plt.imshow(image)
plt.show()
```

Converting the RGB image to Gray Scale

```python

image_gray=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
plt.imshow(image_gray)
```

Inverting the gray scale image

```python

inverted_image= 255-image_gray
plt.imshow(inverted_image)
```

Blurring the image using Gaussian Function

```python

image_blurr=cv2.GaussianBlur(inverted_image, (21, 21),sigmaX=0, sigmaY=0)
plt.imshow(image_blurr)
```

Mixing the Gray-Scale Image with Inverted Blurry Image

```python

def gray_InvBlur(x, y):
    return cv2.divide(x, 255 - y, scale=256)
```

Passing the Gray-Scale and the Iverted Blurry Image to the function for mixing

```python

pencil_sketch_img = gray_InvBlur(image_gray,image_blurr)

# Displaying the Original and the Pencil-Sketch Image
plt.subplots(figsize=(15,10))

plt.subplot(1,2,1)
plt.imshow(image)

plt.subplot(1,2,2)
plt.imshow(pencil_sketch_img)
```

Saving and displaying the Final Pencil Sketch Image

```python
cv2.imwrite('PencilSketch.jpg',pencil_sketch_img)
```
